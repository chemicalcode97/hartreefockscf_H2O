```python
#importing numpy and linalg module from scipy for calculation of matrix propaties
import numpy as np
from scipy import linalg
import pandas as pd
```

```python
def eigenvaluevectorsort(eigenvalues,eigenvectors):
    idx = eigenvalues.argsort()
    eigenvalues=eigenvalues[idx]
    eigenvectors=eigenvectors[:,idx]
    return eigenvalues,eigenvectors
```

```python
E_nuc = np.loadtxt("enuc.txt")
print(E_nuc)
```

```python
#loading the input file for overlap(s), kinetic energy(t),and nuclear-electron(v) integrals
s = np.loadtxt("s.txt")
t = np.loadtxt("t.txt")
v = np.loadtxt("v.txt")
```

```python
#creating some null matrix useful for later calculation
s_matrix = np.zeros(shape=(7,7))
t_matrix = np.zeros(shape=(7,7))
v_matrix = np.zeros(shape=(7,7))
H = np.zeros(shape=(7,7))
```

```python
su = len(s[:,0])
print(su)
```

```python
for a in range(su):
    i = int(s[a][0])
    j = int(s[a][1])
    s_matrix[i-1][j-1]=s[a][2]

S = s_matrix + s_matrix.transpose() - np.diag(np.diag(s_matrix))
print("Overlap Matrix \n",S)   #S is full overlap matrix
```

```python
for a in range(su):
    i = int(t[a][0])
    j = int(t[a][1])
    t_matrix[i-1][j-1]=t[a][2]
T = t_matrix + t_matrix.transpose() - np.diag(np.diag(t_matrix))
print("Kinetic Energy Matrix \n",T)  #T is the full kinetic energy matrix
```

```python
for a in range(su):
    i = int(v[a][0])
    j = int(v[a][1])
    v_matrix[i-1][j-1]=v[a][2]
V = v_matrix + v_matrix.transpose() - np.diag(np.diag(v_matrix))
print("Nuclear-Electron Attraction Matrix \n",V) #V is the full nuclear-electron attraction matrix
```

```python
#constructing Core Hamiltonian
H_core = T + V
print(" Core Hamiltonian \n",H_core)
```

```python
#loading two electron replusion input file
eri = np.genfromtxt("eri.txt")
```

```python
d_su = len(eri[:,0])
print(d_su)
```

```python
eri_matrix = np.zeros(shape=(7,7,7,7))
twoE = np.zeros(shape=(7,7,7,7))
```

```python
for a in range(d_su):
    i = int(eri[a,0]) -1
    j = int(eri[a,1]) -1
    k = int(eri[a,2]) -1
    l = int(eri[a,3]) -1
    
    eri_matrix[i,j,k,l] = eri[a,4]
    twoE[i,j,k,l]=twoE[j,i,k,l]=twoE[i,j,l,k]=twoE[j,i,l,k]=twoE[k,l,i,j]=twoE[l,k,i,j]=twoE[k,l,j,i]=twoE[l,k,j,i]=eri_matrix[i,j,k,l]
    
print("Two-Electron Integrals \n",twoE)
```

```python
#Construct S^(-1/2)

eigen_val , eigen_vec = linalg.eig(S)

eigen_val_mat = np.diag(eigen_val)
eigen_val_mat
eigen_vec_transpose = eigen_vec.transpose()

eigen_val_mat_inv = linalg.inv(eigen_val_mat)
eigen_val_mat_inv_sqroot = linalg.sqrtm(eigen_val_mat_inv)


S_inv_sqroot_f = np.dot(eigen_vec, np.dot(eigen_val_mat_inv_sqroot,eigen_vec_transpose))
S_inv_sqroot = np.round_(S_inv_sqroot_f, decimals = 7 )
S_inv_sqroot = S_inv_sqroot.real
print(" S^(-1/2) \n",S_inv_sqroot)
```

```python
#calculating Density Matrix 

S_inv_sqroot_transpose = S_inv_sqroot.transpose()

F_prime_knot_f = np.dot(S_inv_sqroot_transpose, np.dot(H_core,S_inv_sqroot))
F_prime_knot = np.round_(F_prime_knot_f, decimals= 7)
print("Transformed Fock in The Orthogonal Basis\n",F_prime_knot)

F_prime_knot_eigen_val , F_prime_knot_eigen_vec_f = linalg.eig(F_prime_knot)

F_prime_knot_eigen_vec = np.round_(F_prime_knot_eigen_vec_f, decimals= 7)

F_prime_knot_eigen_val , F_prime_knot_eigen_vec = eigenvaluevectorsort(F_prime_knot_eigen_val , F_prime_knot_eigen_vec)

C_knot_f = np.dot(S_inv_sqroot,F_prime_knot_eigen_vec)
C_knot = np.round_(C_knot_f, decimals= 7)

print("Initial MO Coefficient \n",C_knot) ### 1,3,5,7 doesnt match

density_knot= np.zeros_like(C_knot)

for i in range(7):
    for j in range(7):
        k= 0
        for m in range(5):
            k=k+ C_knot[i,m]*C_knot[j,m]
        density_knot[i,j]= k

print("Initial Density Matrix\n",density_knot)
```

```python
#Calculating Initial Fock Matrix in Original Basis
S_sqroot = linalg.inv(S_inv_sqroot)
F_knot = np.dot(S_sqroot.transpose(),np.dot(F_prime_knot,S_sqroot))
F_knot = np.round_(F_knot,decimals=7)
print("Initial Fock Matrix in Original Basis \n",F_knot)
```

```python
E_elec_knot = 0

for i in range(7):
    for j in range(7):
        E_elec_knot = E_elec_knot + density_knot[i,j]*(H_core[i,j]+F_knot[i,j])
        
E_total = E_elec_knot + E_nuc
E_elec_knot
```

```python
result_matrix = np.array([[E_elec_knot,E_total,"none","none"]])
print(result_matrix)
```

```python
#Building New Fock 
F_1 = np.zeros_like(F_knot)
#F_1
for i in range(7):
        for j in range(7):
            F_1[i,j]=H_core[i,j]
            for k in range(7):
                for l in range(7):
                    F_1[i,j]= F_1[i,j] + density_knot[k,l] * (2*twoE[i,j,k,l]-twoE[i,k,j,l])

F_1 = np.round_(F_1, decimals=7)
print("First iteration Fock Matrix \n",F_1)
```

```python
#calculating First iteration Density Matrix 

F_2_prime = np.dot(S_inv_sqroot_transpose, np.dot(F_1,S_inv_sqroot))
F_2_prime = np.round_(F_2_prime, decimals= 7)
print("Transformed Fock in The Orthogonal Basis\n",F_2_prime)

F_2_prime_eigen_val , F_2_prime_eigen_vec = linalg.eig(F_2_prime)
F_2_prime_eigen_vec = np.round_(F_2_prime_eigen_vec, decimals= 7)
F_2_prime_eigen_val = np.round_(F_2_prime_eigen_val, decimals= 7)
print("F_2_prime_eigen_val \n",F_2_prime_eigen_val)
print("F_2_prime_eigen_vec \n",F_2_prime_eigen_vec)

F_2_prime_eigen_val , F_2_prime_eigen_vec = eigenvaluevectorsort(F_2_prime_eigen_val , F_2_prime_eigen_vec)

print("F_2_prime_eigen_val \n",F_2_prime_eigen_val)
print("F_2_prime_eigen_vec \n",F_2_prime_eigen_vec)

C_1 = np.dot(S_inv_sqroot,F_2_prime_eigen_vec)
C_1 = np.round_(C_1, decimals= 7)
print("First iteration MO Coefficient \n",C_1)


density_1 = np.zeros_like(C_1)

for i in range(7):
    for j in range(7):
        k= 0
        for m in range(5):
            k=k+ C_1[i,m]*C_1[j,m]
        density_1[i,j]= k
density_1 = np.round_(density_1,decimals=7)
print("First iteration density Matrix\n",density_1)
```

```python
#Calculating F_2 Fock Matrix in Original Basis

F_2 = np.dot(S_sqroot.transpose(),np.dot(F_2_prime,S_sqroot))
F_2 = np.round_(F_2,decimals=7)
print("F_2 Fock Matrix in Original Basis \n",F_2)
```

```python
#Calculation of E electronic and E total
E_elec = 0

for i in range(7):
    for j in range(7):
        E_elec = E_elec + density_1[i,j]*(H_core[i,j]+F_2[i,j])
        
E_total = E_elec + E_nuc
E_elec
```

```python
# This is a Function for SCF cycle calculation

def SCFcycle(numberofcycle,density_1,F_2,E_elec,result):
    E_elec_prev = E_elec
    density_prev = density_1
    F_prev = F_2
    
    #count = 2
    
    for z in range(numberofcycle):
        
        #Building New Fock 
        F_init = np.zeros_like(F_prev)
        #F_init
        for i in range(7):
                for j in range(7):
                    F_init[i,j]=H_core[i,j]
                    for k in range(7):
                        for l in range(7):
                            F_init[i,j]= F_init[i,j] + density_prev[k,l] * (2*twoE[i,j,k,l]-twoE[i,k,j,l])

        F_init = np.round_(F_init, decimals=7)
        
        F_New_prime = np.dot(S_inv_sqroot_transpose, np.dot(F_init,S_inv_sqroot))
        F_New_prime = np.round_(F_New_prime, decimals= 7)
        
 
        F_New_prime_eigen_val , F_New_prime_eigen_vec = linalg.eig(F_New_prime)
        F_New_prime_eigen_vec = np.round_(F_New_prime_eigen_vec, decimals= 7)
        F_New_prime_eigen_val = np.round_(F_New_prime_eigen_val, decimals= 7)
 
        F_New_prime_eigen_val , F_New_prime_eigen_vec = eigenvaluevectorsort(F_New_prime_eigen_val , F_New_prime_eigen_vec)

        C_New = np.dot(S_inv_sqroot,F_New_prime_eigen_vec)
        C_New = np.round_(C_New, decimals= 7)
        

        density_New = np.zeros_like(C_knot)

        for i in range(7):
            for j in range(7):
                k= 0
                for m in range(5):
                    k=k+ C_New[i,m]*C_New[j,m]
                density_New[i,j]= k
        density_New = np.round_(density_New,decimals=7)
    
    
        F_New = np.dot(S_sqroot.transpose(),np.dot(F_New_prime,S_sqroot))
        F_New = np.round_(F_New,decimals=7)
    
        E_elec_New = 0

        for ix in range(7):
            for jx in range(7):
                E_elec_New = E_elec_New + density_New[ix,jx]*(H_core[ix,jx]+F_New[ix,jx])
        
        E_total_New = E_elec_New + E_nuc
        

        rmsD = 0
        for iy in range(7):
            for jy in range(7):
                rmsD = rmsD + (density_New[iy,jy]-density_prev[iy,jy])**2
        rmsD = np.sqrt(rmsD)
    
        delta_E = E_elec_New - E_elec_prev
        
        
        if delta_E < 10**(-7) and rmsD < 10**(-7):
            
            result = np.append(result,[[E_elec_New,E_total_New,delta_E,rmsD]],axis=0)
            result = np.append(result,[["converge","0","0","0"]],axis=0)
            break
        else:
            result = np.append(result,[[E_elec_New,E_total_New,delta_E,rmsD]],axis=0)
        
        
        density_prev = density_New
        E_elec_prev = E_elec_New
        
        
    return result
```
